## cleaning the df_copy dataset

## cleaning the problem_copy dataset

```python
problem_copy.isnull().sum()
```

```python
problem_nan_columns = problem_copy.select_dtypes(include=['object']).columns
problem_nan_columns
```

```python
value = 'unknown'
for col in problem_nan_columns:
    problem_copy[col].fillna(value,inplace=True)
```

```python
problem_copy.isnull().sum()
```

```python
imputer = SimpleImputer(strategy='mean')
problem_copy['cap-diameter']= imputer.fit_transform(problem_copy[['cap-diameter']])
problem_copy['stem-height']= imputer.fit_transform(problem_copy[['stem-height']])
```

```python
problem_copy.isnull().sum()
```

```python
# dropping id column in df_copy as its not relevant
df_copy = df_copy.drop("id",axis=1)
```

## merging the 2 datasets to ensure consistent labelling using LabelEncoder

```python
df_copy.head()
```

```python
problem_copy.head()
```

```python
#create another copy of df_copy drop the class column and merge rows with problem_copy
#this is to allow categories from both sets to have the same lablels
df_merge = df_copy.copy()
df_merge = df_merge.drop('class', axis=1)
```

```python
df_merge.head()
```

```python
#drop the id of problem_copy
problem_copy = problem_copy.drop('id',axis=1)
```

```python
problem_copy.head()
```

```python
#store the len of df dataset for splitting after label encoding 
df_len = len(df_merge)
df_len
```

```python
#merging df_merge and problem_copy
merged_df = pd.concat([df_merge,problem_copy], axis=0)
merged_df.head()
```

```python
#check if merged properly
merged_df.shape
```

```python
#label encoding done to turn categorical values to numerical numbers
label_encoder = preprocessing.LabelEncoder()
categorical_cols = merged_df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    merged_df[col] = label_encoder.fit_transform(merged_df[col])
    
```

## Now split up back to df and problem set 

```python
df1_labelled = merged_df.iloc[:df_len]
df2_labelled = merged_df.iloc[df_len:]
```

```python
df1_labelled.shape
```

## Add back the class column from the original df into df1_labelled 

```python
class_column = df['class']
class_column.head()
```

```python
df1_final = pd.concat([class_column, df1_labelled], axis=1)
df1_final.head()
```

## label encode the class to run it through xgboost

```python
df1_final['class']= label_encoder.fit_transform(df1_final['class'])
```

```python
df1_final.head()
# 0 for e and 1 for p
```

```python
y = df1_final.iloc[:,0]
X = df1_final.iloc[:, 1:]
```

```python
#spliting df1_final into test and train set
from sklearn.model_selection import train_test_split 

X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, random_state=0)
```

```python
#further splitting for a validation set 
X_train, X_val, y_train, y_val = train_test_split(X_train_full,y_train_full, random_state=0)
```

```python
pip install xgboost
```

```python
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(n_estimators = 550,
                        learning_rate = 0.5,
                        use_label_encoder = False,
                        eval_metric = "logloss",
                        early_stopping_rounds = 5,
                        n_jobs = -1)
```

```python
xgb_clf.fit(X_train, y_train,                    
            eval_set = [(X_val,y_val)],
            verbose = False)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
pred_test = xgb_clf.predict(X_test)
test_score = accuracy_score(pred_test, y_test)
print("Test score:", np.round(test_score,6))
```

```python
from sklearn.metrics import matthews_corrcoef
mcc = matthews_corrcoef(pred_test,y_test)
print(f"Matthews Correlation Coefficient: {mcc:.6f}")
```

```python
pred_train = xgb_clf.predict(X_train_full)
train_score = accuracy_score(pred_train, y_train_full)
print("Train score:", np.round(train_score,6))
```

```python
mcc = matthews_corrcoef(pred_train,y_train_full)
print(f"Matthews Correlation Coefficient: {mcc:.6f}")
```
